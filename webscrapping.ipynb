```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
project_url= "http://books.toscrape.com/"
```

```python
content= requests.get(project_url)
content.status_code  # 200 statuscode ok  400 not respond  
```

```python
content= content.text
```

```python
doc= bs(content,'html.parser')
doc
```

```python
book_name_tags= doc.find_all('h3')
for i in book_name_tags:
    print(i.find('a')['title'])
```

```python
book_name_tags= doc.find_all('div')
for i in book_name_tags:
    print(i.find('a'))
```

```python
book_prices= doc.find_all('div')
for i in book_prices:
#     print(i.find('p'))
    if i.find('p')!=None:
        print(i.find('p',class_='price_color').text.strip('Â£'))
```

```python
l=[]
book_categories=doc.find_all('li')
# book_categories[0].text
for i in book_categories:
    if(i.find('ul'))!=None:
        print(i.find('ul').text)
```

```python
book_link=[]
book_images=doc.find_all('div',class_ = 'image_container')
for i in book_images:
    book_link.append('http://books.toscrape.com/'+i.find('a')['href'])
print(book_link)
```

```python
for i in book_link:
    book_details = (requests.get(i))
    st1=bs(book_details.text)
    print(st1.find_all('td')[0].text)
```
